```python
import numpy as np
from autolocal import DocumentManager
```

```python
manager = DocumentManager()
bad_columns = [name for name in manager.metadata.columns if 'Unnamed' in name ]
metadata = manager.metadata.drop(columns=bad_columns)
```

```python
metadata['city'].unique()
```

```python
# select cities
cities= [
    'Cupertino',
    'San Jose',
    'Santa Clara',
    'Mountain View',
    'Sunnyvale'    
]

q = {
    'city': cities,
}

idx = np.array([True]*len(metadata), dtype=bool)
for k,v in q.items():
    idx = idx & (np.sum(np.array([metadata[k]==c for c in v]), 0)>0)
meetings = metadata.loc[idx,:]
```

```python
# pull documents
import pandas as pd
from autolocal.nlp import Tokenizer
tokenizer = Tokenizer()

documents = []
for idx, row in meetings.iterrows():
    path = row['local_path_txt']
    with open(path) as f:
        doc = f.read()
    documents.append(str(doc))
    
documents =  pd.DataFrame(documents, index=meetings.index, columns=['text'])
meetings = pd.merge(meetings, documents, right_index=True, left_index=True)
```

```python
import re

out_path = '../out/address_report.html'

# r = r'\d{1,4}[A-Z]?\s([NSEW]\.)?\s(\d{1,3}(st|nd|rd|th))?\s(\w\s)+([A-Z][a-z]{1,3}\.)?'
regex = re.compile('\d{1,6} [\w\s]{1,20}(?:street|st|avenue|ave|road|rd|highway|hwy|square|sq|drive|dr|court|ct|park|parkway|pkwy|circle|cir|boulevard|blvd)\W?(?=\s|$)', re.IGNORECASE)

saving=True
window=100

def write_tag(s, tag='p'):    
    print(s)
    if saving:
        with open(out_path,'a') as f:
            f.write('<{}>'.format(tag) + s +'</{}>'.format(tag))

for idx, row in meetings.iterrows():
    doc = row['text']
    res = regex.search(doc)
#     if res is not None:
#         t0, t1 = res.span()
#         print(doc[int(t0-window/2):int(t1+window/2)])
    if res is not None:
        start = res.start() 
        write_tag('{} {}'.format(row['city'], row['committee']), 'h4')
        write_tag('{}'.format(row['date'], 'h5'))
        row['date']
        write_tag(doc[int(start-window/2):int(start+window/2)], 'p')    
```
